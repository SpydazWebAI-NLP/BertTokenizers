```python
from transformers import BertTokenizer, BertModel

def save_bert_vocab_and_embeddings(model_name="bert-base-multilingual-cased", vocab_file="Bert-Base-Multilingual-cased-vocab.txt", embeddings_file="Bert-Base-Multilingual-embeddings.txt"):
    # Load the BERT model and tokenizer
    model = BertModel.from_pretrained(model_name)
    tokenizer = BertTokenizer.from_pretrained(model_name)

    # Get the vocabulary
    vocabulary = tokenizer.get_vocab()

    # Save vocabulary to file
    with open(vocab_file, 'w', encoding='utf-8') as vocab_file:
        for token, index in vocabulary.items():
            vocab_file.write(f"{token}: {index}\n")

    # Save embeddings to file
    embeddings = model.get_input_embeddings().weight.detach().numpy()
    with open(embeddings_file, 'w', encoding='utf-8') as embeddings_file:
        for token, index in vocabulary.items():
            # Combine token, index, and embedding values
            embedding_values = " ".join(map(str, embeddings[index]))
            embeddings_file.write(f"{token} {embedding_values}\n")

if __name__ == "__main__":
    # Replace "bert-base-uncased" with the specific model name if you're using a different one
    save_bert_vocab_and_embeddings(model_name="bert-base-multilingual-cased", vocab_file="bert_base_multilingual-cased-vocab.txt", embeddings_file="bert_base_Multilingual-Cased_embeddings.txt")
```
